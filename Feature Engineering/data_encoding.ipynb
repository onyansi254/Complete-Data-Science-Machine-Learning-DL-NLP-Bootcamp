## Nominal/OHE Encoding
One hot encoding also know as nominal encoding is a technique used to represent categorical data as numerica data, which is moe suitable for machine learning algorithms. In ths technique, each category is represented as binary vector where each bit corresponds to unique category. For example if we have a categorical variable "color" with three possible vaues(red green, bleue) we cn represent it one encoding as folows:
1.Red: [1,0,0]
2.Green:[0,1,0]
3.Blue:[0,0,1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## create a simple dataframe
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red']
})
```

```python
df.head()
```

```python
## create the OneHotEncoder instance
encoder = OneHotEncoder()
```

```python
## perform fit then transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```
