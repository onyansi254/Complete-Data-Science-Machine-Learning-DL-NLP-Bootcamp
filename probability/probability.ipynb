Here‚Äôs a simple explanation of both **Addition Rule** and **Multiplication Rule** in probability, covering **mutual/non-mutual exclusive** events and **independent/dependent** events:
---

### üéØ **1. Addition Rule (For Mutually & Non-Mutually Exclusive Events):**

This rule is used to find the probability of **either** of two events happening.

#### ‚úÖ **Mutually Exclusive Events:**

* Events that **cannot happen at the same time**.
* Example: Tossing a coin ‚Äî getting heads **or** tails (you can't get both at once).
* **Formula:**

$$
P(A \cup B) = P(A) + P(B)
$$

#### ‚úÖ **Non-Mutually Exclusive Events:**

* Events that **can happen together**.
* Example: Drawing a card ‚Äî event A: drawing a king; event B: drawing a heart (king of hearts overlaps).
* **Formula:**

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

Where $P(A \cap B)$ is the probability that both events occur.

---

### üéØ **2. Multiplication Rule (For Independent & Dependent Events):**

This rule is used to find the probability of **both** events happening together.

#### ‚úÖ **Independent Events:**

* Events where the outcome of one does **not affect** the other.
* Example: Tossing two different coins.
* **Formula:**

$$
P(A \cap B) = P(A) \times P(B)
$$

#### ‚úÖ **Dependent Events:**

* Events where the outcome of one **affects** the other.
* Example: Drawing two cards **without replacement**.
* **Formula:**

$$
P(A \cap B) = P(A) \times P(B \mid A)
$$

Where $P(B \mid A)$ is the **conditional probability** of B happening given A has already happened.
