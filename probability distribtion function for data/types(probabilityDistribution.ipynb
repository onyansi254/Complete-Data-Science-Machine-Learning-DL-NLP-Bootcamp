## ‚úÖ **1. Bernoulli Distribution**

* **Scenario:** Single trial with 2 outcomes (success/failure).
* **PMF:**

$$
P(X=x)=p^x(1-p)^{1-x},\;x\in\{0,1\}
$$

* **Mean:** $E[X]=p$
* **Variance:** $\mathrm{Var}(X)=p(1-p)$

**Example Q:** A coin has $P(\text{Heads})=0.7$. What is $P(\text{Heads})$?

$$
P(X=1)=0.7
$$

## ‚úÖ **2. Binomial Distribution**

* **Scenario:** Number of successes in **n independent Bernoulli trials**.
* **PMF:**

$$
P(X=k)=\binom{n}{k}p^k(1-p)^{n-k},\;k=0,1,\dots,n
$$

* **Mean:** $E[X]=np$
* **Variance:** $\mathrm{Var}(X)=np(1-p)$

**Example Q:** Toss a fair coin $3$ times. What‚Äôs $P(\text{exactly 2 heads})$?

$$
P(X=2)=\binom{3}{2}(0.5)^2(0.5)=0.375
$$

## ‚úÖ **3. Geometric Distribution**

* **Scenario:** Number of trials until the **first success**.
* **PMF:**

$$
P(X=k)=(1-p)^{k-1}p,\;k=1,2,\dots
$$

* **Mean:** $E[X]=\frac{1}{p}$
* **Variance:** $\mathrm{Var}(X)=\frac{1-p}{p^2}$

**Example Q:** Player scores with $p=0.4$. What‚Äôs $P(\text{success on 3rd trial})$?

$$
P(X=3)=0.6^2\cdot0.4=0.144
$$

## ‚úÖ **4. Negative Binomial Distribution**

* **Scenario:** Number of trials needed for **r successes**.
* **PMF:**

$$
P(X=k)=\binom{k-1}{r-1}p^r(1-p)^{k-r},\;k=r,r+1,\dots
$$

* **Mean:** $E[X]=\frac{r}{p}$
* **Variance:** $\mathrm{Var}(X)=\frac{r(1-p)}{p^2}$

**Example Q:** Coin with $p=0.5$. What‚Äôs $P(\text{3rd head on 5th toss})$?

$$
P(X=5)=\binom{4}{2}(0.5)^3(0.5)^2=0.15625
$$
