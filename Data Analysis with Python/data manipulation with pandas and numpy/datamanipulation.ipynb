```python
grouped_mean_volume = df.groupby('Symbol')['Volume'].mean()
grouped_mean_volume
```

```python
### Mergind=g and joining dataframe
# create sample dataframes
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': [1,2,3,4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value': [5,6,7,8]})
```
