### Label Encoding

Label encoding and ordinal encoding are two techniques used to encode categorical data as numerical data. 

label encoding involved assigning  numerical abel to each category in the variable. The labels are usualy assigned in alphabetic order or based on the freqency of the categories. For example, if we have a categorical variabe "color" with three possible values(red,green,blue), we can represent it using label encoding as follows:
1. Red:1
2. Green:2
3. Blue:3

```python
import pandas as pd
#from sklearn.preprocessing import OneHotEncoder
```

```python
## create a simple dataframe
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red']
})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.fit_transform([['red']])
```

```python
## Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
# create a simple  df
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
df
```

```python
## create an instance of OrdinalEncoder and the fit_transform the data
encoder=OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])
```
