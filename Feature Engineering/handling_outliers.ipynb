```python
# minimum,maimum,median,quartiles(25th,75th),IQR
```

```python
import numpy as np
```

```python
lst_marks=[23,45,67,89,12,34,56,78,90,11,22,33,44,55,66,77,88,99,100,5,3,2,1,0,150,200,250] 
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1])
```

```python
minimum, Q1, median, Q3, maximum
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
upper_fence=Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
upper_fence
```
