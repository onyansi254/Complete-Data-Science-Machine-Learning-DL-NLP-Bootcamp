Here are the **formulas for measures of dispersion** for both **population** and **sample data**:

---

### **1. Variance (σ² and s²)**

| Type                         | Formula                                      | Description                         |
| ---------------------------- | -------------------------------------------- | ----------------------------------- |
| **Population Variance (σ²)** | $\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}$    | Divide by total population size $N$ |
| **Sample Variance (s²)**     | $s^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1}$ | Divide by degrees of freedom $n-1$  |

---

### **2. Standard Deviation (σ and s)**

Simply the square root of variance:

| Type                  | Formula                    |
| --------------------- | -------------------------- |
| **Population SD (σ)** | $\sigma = \sqrt{\sigma^2}$ |
| **Sample SD (s)**     | $s = \sqrt{s^2}$           |

---

### **Key Differences:**

* **Population formulas** use **N** (all data points), since it's the entire group.
* **Sample formulas** use **n-1** (Bessel's correction) to correct bias when estimating population from a sample.

---

### **Other Common Measures of Dispersion:**

| Measure                           | Formula                   |            |                                     |                |                 |
| --------------------------------- | ------------------------- | ---------- | ----------------------------------- | -------------- | --------------- |

| **Mean Absolute Deviation (MAD)** | ( \frac{\sum              | x\_i - \mu | }{N} ) (population) or ( \frac{\sum | x\_i - \bar{x} | }{n} ) (sample) |
| **Interquartile Range (IQR)**     | $Q_3 - Q_1$               |            |                                     |                |                 |

### ✅ **Why Do We Divide by $n - 1$ Instead of $n$ in Sample Variance?**

This is due to **Bessel’s Correction**. It corrects the bias in the estimation of the population variance from a sample.

#### Reason:

* In sample variance, we use the **sample mean ($\bar{x}$)** instead of the **true population mean ($\mu$)**.
* The sample mean itself is calculated from the same data, which tends to **underestimate** the population variance.
* Dividing by $n - 1$ instead of $n$ compensates for this bias, giving a more accurate estimate of the population variance.

#### In simple terms:

* $n - 1$ adjusts for the fact that we've already "used up" one degree of freedom by calculating the sample mean.
* This ensures that the variance isn’t underestimated.

---

### ✅ **Possible Interview Questions:**

1. **Why do we use $n - 1$ instead of $n$ in the sample variance formula?**
   *Expected Answer:*
   To correct for bias caused by using the sample mean instead of the population mean, ensuring an unbiased estimate of the population variance (Bessel’s Correction).

2. **Explain the difference between population variance and sample variance.**
   *Expected Answer:*
   Population variance uses the actual population mean and divides by $N$, while sample variance uses the sample mean and divides by $n - 1$ to adjust for bias.

3. **What is meant by "degrees of freedom" in statistics?**
   *Expected Answer:*
   Degrees of freedom refer to the number of independent values that can vary in an analysis. In sample variance, we subtract 1 because one value (the mean) is already fixed.

4. **When would you use sample variance instead of population variance?**
   *Expected Answer:*
   When you're working with a sample and trying to estimate the variance of the larger population.

5. **What happens if you mistakenly divide by $n$ instead of $n - 1$ for sample variance?**
   *Expected Answer:*
   You’d underestimate the true population variance because you wouldn’t account for the bias introduced by estimating the mean from the sample itself.

### ✅ **Standard Deviation (SD):**

**Standard deviation tells how far, on average, each data point is from the mean.**

* A **small SD** means data points are **close to the mean** (less spread out).
* A **large SD** means data points are **more spread out** (greater variability).

#### Formula:

* **Population SD:**

$$
\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}
$$

* **Sample SD:**

$$
s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}}
$$

#### In Simple Words:

> Standard deviation measures the **typical distance** from the **mean**.
> It shows **how consistent or variable** the data is.

---

### ✅ **Example (Quick View):**

\| Data: | 2, 4, 4, 4, 5, 5, 7, 9 |
\| Mean: | 5 |
\| SD:   | 2 |

> Interpretation:
> On average, numbers in the dataset are about **2 units away** from the mean (5).

---

### ✅ **Sample Interview Question:**

**Q:** What does standard deviation tell you about a dataset?
**A:** It tells me how spread out the data points are around the mean. A smaller standard deviation means the data points are closely clustered, while a larger standard deviation means they are more spread out.

In **statistics**, a **variable** refers to any characteristic, number, or quantity that can be measured or counted. Variables represent the properties of individuals or objects under study, and they can change from one entity to another or over time.

---

### ✅ **Types of Variables in Statistics:**

Variables are broadly categorized into **two major types**:

1. **Qualitative (Categorical) Variables**
2. **Quantitative (Numerical) Variables**

---

### 1. **Qualitative (Categorical) Variables**

These variables describe **qualities** or **categories** and cannot be measured numerically. They represent groups or labels.

#### a. **Nominal Variables**

* No inherent order or ranking.
* Used for labeling categories.
* Examples:

  * Gender (Male, Female)
  * Blood Type (A, B, AB, O)
  * Colors (Red, Green, Blue)

#### b. **Ordinal Variables**

* Categories with a **meaningful order** or ranking, but differences between categories may not be uniform or measurable.
* Examples:

  * Education Level (High School, Bachelor’s, Master’s, PhD)
  * Satisfaction Level (Very Unsatisfied, Unsatisfied, Neutral, Satisfied, Very Satisfied)
  * Customer Ratings (1-star, 2-star, 3-star)

---

### 2. **Quantitative (Numerical) Variables**

These variables are **measurable** and expressed in numbers.

#### a. **Discrete Variables**

* Can take **specific, distinct values** (often whole numbers).
* Usually results from **counting**.
* Examples:

  * Number of students in a class
  * Number of cars in a parking lot
  * Number of goals scored

#### b. **Continuous Variables**

* Can take **any value** within a given range.
* Usually results from **measuring**.
* Examples:

  * Height (cm)
  * Weight (kg)
  * Temperature (°C)
  * Time (seconds)

---

### ✅ **Other Common Classifications (Advanced Use):**

| Type                     | Description                                                                                | Examples                                                 |
| ------------------------ | ------------------------------------------------------------------------------------------ | -------------------------------------------------------- |
| **Independent Variable** | Variable manipulated or categorized to observe its effect                                  | Dose of medicine in an experiment                        |
| **Dependent Variable**   | Variable observed or measured to see the effect                                            | Patient's recovery rate                                  |
| **Confounding Variable** | Hidden variable affecting both independent & dependent variables, possibly biasing results | Age affecting both exercise frequency and blood pressure |
| **Control Variable**     | Kept constant to prevent influence on results                                              | Room temperature in an experiment                        |

---

### ✅ **Summary Table:**

| Type                          | Subtype    | Description                                    | Examples                             |
| ----------------------------- | ---------- | ---------------------------------------------- | ------------------------------------ |
| **Qualitative (Categorical)** | Nominal    | Categories without order                       | Gender, Colors, Blood Group          |
|                               | Ordinal    | Categories with meaningful order               | Education Level, Satisfaction Rating |
| **Quantitative (Numerical)**  | Discrete   | Countable numbers, often whole numbers         | Number of Cars, Students             |
|                               | Continuous | Measurable numbers with infinite possibilities | Height, Weight, Speed                |

---

### ✅ **Key Takeaway:**

* **Qualitative = Categories (labels or groups)**
* **Quantitative = Numbers (counts or measurements)**

In **statistics**, a **random variable** is like a **function** that takes the **outcomes** (or **results**) of an **experiment** or **process** and gives a **value** (usually a number).

➡️ You do an **experiment** or a **random process** (like rolling a dice).

➡️ The **random variable** is the **function** that assigns a **number** to each possible result.

### Example:

* **Experiment/process:** Toss a coin.
* **Random variable (function):** Assign 1 for heads, 0 for tails.
* **Values:** 0 or 1 (depending on result).

---

In short:
**Random variable = function → takes experiment → gives values.**
