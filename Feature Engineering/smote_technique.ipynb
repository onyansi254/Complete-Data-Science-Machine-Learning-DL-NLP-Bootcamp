### SMOTE (Synthetic Minority Oversampling Technique)

```python
# smote is a technique used in machine learning to address imbalaced datasets by generating synthetic samples for the minority class.
# SMOTE involves generating synthtc instances by interpolating between existing minority class samples and their nearest neighbors.
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                    weights=[0.9],random_state=12)
```

```python
import pandas as pd

df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset using SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```
