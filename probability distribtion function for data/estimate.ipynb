## âœ… 1. What is an Estimate?

* An **estimate** is a value or range of values derived from sample data to **approximate an unknown population parameter** (like mean, variance, or proportion).
* For example:

  * You canâ€™t measure the **true average height** of all adults in Kenya, so you take a sample and **estimate** the mean.

## âœ… 2. Types of Estimates

### **A) Point Estimate**

* A **single number** that serves as the best guess for the population parameter.
* Example:

  * Sample mean ($\bar{x}$) is a **point estimate** of the population mean ($\mu$).
  * Sample proportion ($\hat{p}$) is a **point estimate** of the population proportion ($p$).

ðŸ“Œ **Properties of a Good Point Estimator**

* **Unbiased** â†’ Expected value equals the true parameter
* **Consistent** â†’ Improves with larger samples
* **Efficient** â†’ Has the smallest variance among all unbiased estimators

---

### **B) Interval Estimate (Confidence Interval)**

* Instead of giving one number, you give a **range** likely to contain the true parameter.
* Example:

  * You might say: *The average height is **170 cm Â± 2 cm** with 95% confidence.*

**Formula for a Confidence Interval (CI):**

$$
\text{Point Estimate} \pm \text{Margin of Error}
$$

* Margin of Error depends on:

  * Confidence level (e.g. 95%)
  * Standard error of the sample

Example for population mean (when Ïƒ is known):

$$
\bar{x} \pm Z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
$$

---

### **C) Other Special Types**

1. **Maximum Likelihood Estimate (MLE)**

   * Chooses the parameter value that **maximizes the likelihood** of observing the given data.
2. **Bayesian Estimate**

   * Uses **prior knowledge** + sample data to estimate parameters.

---

## âœ… 3. Example

* You want to estimate the **average exam score** of students in a university.

* You take a sample of **50 students**, find the **sample mean = 75**, and **sample standard deviation = 10**.

* **Point Estimate:**
  $\hat{\mu} = 75$

* **95% Confidence Interval:**

  $$
  75 \pm 1.96 \left( \frac{10}{\sqrt{50}} \right) = 75 \pm 2.77
  $$

  So the **CI** is **(72.23, 77.77)**.

## âœ… 1. POINT ESTIMATES

### **1.1 Common Point Estimators**

| **Population Parameter**         | **Point Estimator**           | **Formula**                                |
| -------------------------------- | ----------------------------- | ------------------------------------------ |

### âœ… **Example 1: Point Estimate of Mean**

### **2.1 Confidence Interval for the Population Mean**

* $Z_{\alpha/2}$ = critical value from standard normal
* $t_{\alpha/2, n-1}$ = critical value from t-table

## âœ… 3. OTHER ESTIMATES

---

### **3.1 Maximum Likelihood Estimate (MLE)**

* MLE finds the parameter value that **maximizes the likelihood function**.
* Example: For a **Bernoulli distribution**, the MLE of $p$ is just $\hat{p} = \frac{\text{successes}}{n}$.

---

### **3.2 Bayesian Estimate**

* Bayesian estimation **combines prior belief (prior distribution)** with **sample data** to get a **posterior estimate**.
* Used when you have **prior knowledge**.
