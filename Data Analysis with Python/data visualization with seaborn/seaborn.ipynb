```python
# seaborn is a library for data visualization
#seaborn helps in creating comlex visualization with just few lines of codes.
```

```python
import seaborn as sns
```

```python
# Basic plotting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
# create scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
## line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Total Bill vs Size')
plt.xlabel('Size')
plt.ylabel('Total Bill')
plt.show()
```

```python
## categorical plots
## BAR PLOT
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
## Box plot - used to see outliers
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
## violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.grid(True)
plt.show()
```

```python
## histogram 
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
## KDE PLOT
sns.kdeplot(tips['total_bill'] , shade=True, fill=True)
```

```python
# pairplot
sns.pairplot(tips)
```

```python
## HEAT MAP
corr = tips[['total_bill','tip','size']].corr()
corr
#sns.heatmap(corr, annot=True)
#plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
